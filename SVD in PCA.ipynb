```python
import pandas as pd
import numpy as np
np.set_printoptions(suppress=True)
```

```python
FoodRating=pd.read_csv("C:/Subhajit/Newfolder/IOT/DS/Machine Learining/PCA/MyFoodRatings.csv")
FoodRating.head()
```

```python
# Drop all non-numeric values
FoodRating.drop('Name',1,inplace=True)
```

```python
#Covariance matrix
C=np.cov(FoodRating.T)
```

```python
#eigenvalues decomposition of the matrix
eigenvalues,eigenvectors=np.linalg.eig(C)
```

```python
#lets sort them
idx=eigenvalues.argsort()[::-1]
eigenvalues=eigenvalues[idx]
eigenvectors=eigenvectors[:,idx]
```

```python
eigenvalues
```

```python
eigenvectors
```

### Using SVD

```python
# use the SVD function from np.linalg library
U,s,VT=np.linalg.svd(FoodRating,full_matrices=True)
```

```python
#Let's denote the eigenvectors matrix as X. This is same as the principal componets
X=eigenvectors
X
```

```python
# Projecting the data now by doing  the basis transformation
datn=np.linalg.inv(X) @ FoodRating.T
datn.T
```
